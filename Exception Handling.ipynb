# Exception Handling in Python

## Definition
Exception Handling is a mechanism in Python used to handle runtime errors and prevent the program from terminating unexpectedly.

---

## Advantages of Exception Handling

- Prevents program crashes.
- Maintains normal program flow.
- Makes code more reliable and readable.
- Helps in debugging errors.

---

## Common Exceptions in Python

| Exception | Description |
|------------|-------------|
| ZeroDivisionError | Division by zero |
| ValueError | Invalid value |
| TypeError | Wrong data type |
| IndexError | Invalid list index |
| KeyError | Invalid dictionary key |
| NameError | Variable not defined |
| FileNotFoundError | File not found |

---

## try and except

```python
try:
    result = 10 / 0
except ZeroDivisionError:
    print("Cannot divide by zero")
```

### Output

```text
Cannot divide by zero
```

---

## Multiple Exceptions

```python
try:
    num = int(input("Enter a number: "))
    result = 10 / num

except ValueError:
    print("Invalid Input")

except ZeroDivisionError:
    print("Division by Zero Not Allowed")
```

---

## try-except-else

The `else` block executes when no exception occurs.

```python
try:
    num = int(input("Enter a number: "))
except ValueError:
    print("Invalid Input")
else:
    print("You entered:", num)
```

---

## try-except-finally

The `finally` block always executes whether an exception occurs or not.

```python
try:
    num = 10 / 2
except ZeroDivisionError:
    print("Error")
finally:
    print("Execution Completed")
```

---

## Raising Exceptions

The `raise` keyword is used to generate an exception manually.

```python
age = -5

if age < 0:
    raise ValueError("Age cannot be negative")
```

---

## User Defined Exception

```python
class MyException(Exception):
    pass

try:
    raise MyException("Custom Exception")
except MyException as e:
    print(e)
```

---

## Complete Example

```python
try:
    a = int(input("Enter first number: "))
    b = int(input("Enter second number: "))

    result = a / b

except ValueError:
    print("Please enter valid numbers")

except ZeroDivisionError:
    print("Division by zero is not allowed")

else:
    print("Result =", result)

finally:
    print("Program Ended")
```

---

## Exception Handling Flow

```text
try
 │
 ├── No Exception ──► else
 │
 ├── Exception ─────► except
 │
 ▼
finally
```
