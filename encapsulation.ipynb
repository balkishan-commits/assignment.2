## Encapsulation 

```python
class Company:
    company_name = "rangers pvt ltd".  ### public variable
    _company_add  = "jaipur".           ### public variable
    __account no. = 23456765432.       ### private variable

def show(self):
    print(self.__account no.)

dev = Company
dev.show()    
```
